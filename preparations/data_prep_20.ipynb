```python
#Dependencies
import pandas as pd
import numpy as np
from sportsreference.nba.schedule import Schedule
from datetime import date
```

```python
#bucks schedule object
bucks_schedule_20 = Schedule('MIL', 2020)
```

```python
#bucks data frame *focusing on pre-bubble regular season game"
dates = []
results = []
locations = []
game_nums = []
opponents = []

for game in bucks_schedule_20:
    
    if game.playoffs == False and game.game < 66:
        print(game.datetime)  
        print(game.result) 
        print(game.location)
        print(game.game)
        dates.append(game.datetime)
        results.append(game.result)
        locations.append(game.location)
        game_nums.append(game.game)
        opponents.append(game.opponent_abbr)
```

```python
#creating bucks dictionary
bucks_dict = {
    "dates": dates,
    "results": results,
    "locations": locations,
    "game_nums": game_nums,
    "opponent": opponents
}

#creating bucks data frame
bucks_df = pd.DataFrame(bucks_dict)
bucks_df
```

```python
#finding the number of games between each game
bucks_df["date_diff"] = 0

for i in range(1,len(bucks_df)):    
    bucks_df["date_diff"][i] = bucks_df["dates"][i] - bucks_df["dates"][(i-1)]
    
bucks_df
```

```python
#inserting value for the date diff in the first game
#using the difference between the last preseason game and the first regular season game
bucks_df["date_diff"][0] = 5

#converting to a time delta
bucks_df["date_diff"][0] = pd.to_timedelta(bucks_df["date_diff"][0], unit = "D")


bucks_df
```

```python
#Converting the date difference series back into an integer for easier comparisons
for day in range(0,len(bucks_df["date_diff"])):
    bucks_df["date_diff"][day] = bucks_df["date_diff"][day] / np.timedelta64(1, 'D')


bucks_df
```

```python
#date diff is only the difference between the two games
#creating a column to represent the "rest" days between the games
#"rest" in this case is the days between each game where a game is not being played and does not account for practices
bucks_df["rest"] = bucks_df["date_diff"] - 1 

bucks_df.head()
```

```python
#write to csv
bucks_df.to_csv("../data/bucks_19_20_rest.csv", index = False)
```
