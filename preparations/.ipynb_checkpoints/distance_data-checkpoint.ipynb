```python
#dependencies
import pandas as pd
from geopy import distance
```

```python
#setting the stadium coordinates, key is the 3 letter team code
arenas_dict = {
    'MIL': (43.0451,  87.9174),
    'TOR': (43.6435,  79.3791),
    'BOS': (42.3662, 71.0621),
    'MIA': (25.7814,  80.1870),
    'IND': (39.7640,  86.1555), 
    'PHI': (39.9012, 75.1720), 
    'BRK': (40.6826, 73.9754), 
    'ORL': (28.5392, 81.3839), 
    'WAS': (38.8981, 77.0208), 
    'CHO': (35.2252, 80.8394), 
    'CHI': (41.8807, 87.6742), 
    'NYK': (40.7505, 73.9934), 
    'DET': (42.3411, 83.0553), 
    'CLE': (41.4966, 81.6881), 
    'ATL': (33.7573, 84.3963), 
    'LAL': (34.0430, 118.2673), 
    'LAC': (34.0430, 118.2673), 
    'DEN': (39.7487, 105.0076), 
    'UTA': (40.7683, 111.9011), 
    'OKC': (35.4634, 97.5151), 
    'HOU': (29.7508, 95.3621), 
    'DAL': (32.7903, 96.8101), 
    'MEM': (35.1381, 90.0506), 
    'POR': (45.5316, 122.6668), 
    'NOP': (33.7114, 78.8844), 
    'SAC': (38.5802, 121.4997), 
    'SAS': (29.4272, 98.4375), 
    'PHO': (33.4457, 112.0712), 
    'MIN': (44.9795, 93.2761), 
    'GSW': (37.7680, 122.3877), 
}
```

```python
#importing the combined years data
bucks_df = pd.read_csv("../data/bucks_16_20_rest.csv")
bucks_df
```

```python
#putting arena into the dataframe

bucks_df["game_arenas"] = ""

for i in range(0,len(bucks_df)):
    if bucks_df["locations"][i] == "Home":
        bucks_df["game_arenas"][i] = arenas_dict.get("MIL")
    else:
        for key in arenas_dict:
            if key == bucks_df["opponent"][i]:
                bucks_df["game_arenas"][i] = arenas_dict.get(key)
         
```

```python
bucks_df
```

```python
#logic psuedo code for distance traveled in miles 

bucks_df["travel_miles"] = 0

for i in range(0,len(bucks_df)):
    
    if bucks_df["game_nums"][i] == 1 and bucks_df["locations"][i] == "Home":
        bucks_df["travel_miles"][i] = 0
        
    elif bucks_df["game_nums"][i] == 1 and bucks_df["locations"][i] == "Away":
        bucks_df["travel_miles"][i] = distance.distance(arenas_dict["MIL"], bucks_df["game_arenas"][i]).miles
```

```python
bucks_df.head(166)
```

```python
#now the bigger logic

for i in range(0,len(bucks_df)):
    #making sure to not overwrite the game 1 data
    if bucks_df["game_nums"][i] != 1:
        
        #if the previous game was home and they're staying home there is no travel
        if bucks_df["locations"][i-1] == "Home" and bucks_df["locations"][i] == "Home":
            bucks_df["travel_miles"][i] = 0
        
        #if the previous game was home and they're traveling away calculate the distance
        elif bucks_df["locations"][i-1] == "Home" and bucks_df["locations"][i] == "Away":
            bucks_df["travel_miles"][i] = distance.distance(arenas_dict["MIL"], bucks_df["game_arenas"][i]).miles
          
        #if the previous game was away and the current game is home calculate the distance
        elif bucks_df["locations"][i-1] == "Away" and bucks_df["locations"][i] == "Home":
            bucks_df["travel_miles"][i] = distance.distance(arenas_dict["MIL"], bucks_df["game_arenas"][i-1]).miles
            
        #if the previous game was away and the current game is away claculate the distance
        elif bucks_df["locations"][i-1] == "Away" and bucks_df["locations"][i] == "Away":
            bucks_df["travel_miles"][i] = distance.distance(bucks_df["game_arenas"][i], bucks_df["game_arenas"][i-1]).miles
        

        
        
```

```python
#writing the new distance data to csv
bucks_df.to_csv("../data/bucks_16_20_rest_dist.csv", index = False)
```
